# Adding Label/Features

- We adding label - what we want that the model'll predict. 
- The features that help the model predict and understend the data better

```python
window = 50
```

```python
df = data.copy()

## Create features 

# market direction (1 up, 0 down)
df["dir"] = np.where(df["returns"] > 0, 1, 0)

# simple moving avarage (but the distance between sma 50 to sma 150) 
# its create a feature as cross SMA strategy.
df["sma"] = df[symbol].rolling(window).mean() - df[symbol].rolling(150).mean()

# Feature as Bolinnger bounds
#  (price - rolling_avarage_price) / rolling_std 
# in other words, how many deviation (+/-) the price far from avarage ? 
df["boll"] = (df[symbol] - df[symbol].rolling(window).mean()) / df[symbol].rolling(window).std()

# rolling min/max
df["min"] = df[symbol].rolling(window).min() / df[symbol] - 1
df["max"] = df[symbol].rolling(window).max() / df[symbol] - 1

# rolling avarage of returns (momentume feature) 
df["mom"] = df["returns"].rolling(3).mean()

# rolling volatility of returns (the std.) 
df["vol"] = df["returns"].rolling(window).std()

# Drop NaN
df.dropna(inplace = True)
```

```python
df
```

# Adding Feature lags

```python
lags = 5
```

```python
cols = []
# features to lag
features = ["dir", "sma", "boll", "min", "max", "mom", "vol"]
```

```python

# For each feature
for f in features: 
        # for each lag
        for lag in range(1, lags + 1):
            # create the lags 
            col = "{}_lag_{}".format(f, lag)
            df[col] = df[f].shift(lag)
            cols.append(col)
df.dropna(inplace = True)
```

```python
# 1. short where proba < 0.47
test["position"] = np.where(test.proba < 0.47, -1, np.nan) 
```

```python
# 2. long where proba > 0.53
test["position"] = np.where(test.proba > 0.53, 1, test.position) 
```

```python
# Trade in busy hours

# Convert time to UTC
test.index = test.index.tz_localize("UTC")
# Convert & save NY time
test["NYTime"] = test.index.tz_convert("America/New_York")
# stire the hours 
test["hour"] = test.NYTime.dt.hour
```

```python
# 3. neutral in non-busy hours
test["position"] = np.where(~test.hour.between(2, 12), 0, test.position) 
```

```python
# 4. in all other cases: hold position
test["position"] = test.position.ffill().fillna(0) 
```

```python
test.position.value_counts(dropna = False)
# 0.00000     11604: 11604 times hold 
# -1.00000     6436: 6436 times take short
# 1.00000       875: 875 times take long
```

```python
# Calculate the returns of the strategy
test["strategy"] = test["position"] * test["returns"]
```

```python
# Cumelative return of buy and hold / DNN strategy
test["creturns"] = test["returns"].cumsum().apply(np.exp)
test["cstrategy"] = test["strategy"].cumsum().apply(np.exp)
```

```python
# Plot 
test[["creturns", "cstrategy"]].plot(figsize = (12, 8))
plt.show()
```

# IBKR Implementation  

```python
# import libraries
from ib_insync import * 
import pandas as pd
import numpy as np
from IPython.display import display, clear_output
from datetime import datetime, timedelta
import time
import datetime as dt
util.startLoop()
```

```python
# Create an instance & connect
ib = IB()
ib.connect()
```

```python
# strategy parameters
freq = "1 min"
units = 1000
end_time = dt.time(15, 56, 0) # stop condition 15:54:00
contract = Forex('EURUSD') 
ib.qualifyContracts(contract)
cfd = CFD("EUR", currency = "USD")
ib.qualifyContracts(cfd)
conID = cfd.conId
```

```python
dt.datetime.utcnow().time()
```

```python
dt.datetime.utcnow().time() >= end_time
```

```python
session_start = pd.to_datetime(dt.datetime.utcnow()).tz_localize("utc")
```

```python
symbol = "EURUSD"
window = 150
```

__Driven data + run the strategy__ 

```python
def onBarUpdate(bars, hasNewBar):  
    global df, last_bar
    
    if bars[-1].date > last_bar: 
        last_bar = bars[-1].date
    
        # --------------------- Data Processing ------------------------
        df = pd.DataFrame(bars)[["date", "close"]].iloc[:-1] 
        df[symbol] = df["close"]
        df.dropna(inplace=True)
        df["returns"] = np.log(df[symbol] / df[symbol].shift()) 
        df.dropna(inplace=True)
        df.set_index("date", inplace = True)
        # ---------------------------------------------------------------
        
        # ======================== Trading Strategy ===========================
    
        # #-------------------------- Features ------------------------------#
        
        df["dir"] = np.where(df["returns"] > 0, 1, 0) # market direction (1 up, 0 down)
        df["sma"] = df[symbol].rolling(window).mean() - df[symbol].rolling(150).mean()
        df["boll"] = (df[symbol] - df[symbol].rolling(window).mean()) / df[symbol].rolling(window).std()
        df["min"] = df[symbol].rolling(window).min() / df[symbol] - 1
        df["max"] = df[symbol].rolling(window).max() / df[symbol] - 1
        df["mom"] = df["returns"].rolling(3).mean()
        df["vol"] = df["returns"].rolling(window).std()
        df.dropna(inplace = True)

        # create lags Features
        cols = []
        features = ["dir", "sma", "boll", "min", "max", "mom", "vol"]
        for f in features:
            for lag in range(1, 5 + 1):
                col = "{}_lag_{}".format(f, lag)
                df[col] = df[f].shift(lag)
                cols.append(col)
        df.dropna(inplace = True)
       
        df_s = (df - mu) / std # standardization

        # --------------------------------------------------------------------------- #



        #  ---------------------------------prediction-------------------------------# 
        
        df["proba"] = model.predict(df_s[cols])        

        # determine positions
        df = df.loc[session_start:].copy() # starting with first live_stream bar (removing historical bars)
        df["position"] = np.where(df.proba < 0.47, -1, np.nan)
        df["position"] = np.where(df.proba > 0.53, 1, df.position)
        df["position"] = df.position.ffill().fillna(0) # start with neutral position if no strong signal
        
        # --------------------------------------------------------------------------- #

        # =================================================================================
        
        # ---------------------------------- Trading ----------------------------------
        if(len(df) > 0):
            target = df["position"][-1] * units
            execute_trade(target = target)
        
        # Display
        clear_output(wait=True)
        display(df)
    else:
        try:
            trade_reporting()
        except:
            pass

    # ---------------------------------------------------------------------------------#
```

__Funtion to execute a trade and reporting__

```python
def execute_trade(target):
    global current_pos
    
    # 1. get current Position
    try:
        current_pos = [pos.position for pos in ib.positions() if pos.contract.conId == conID][0]
    except:
        current_pos = 0
         
    # 2. identify required trades
    trades = target - current_pos
        
    # 3. trade execution
    if trades > 0:
        side = "BUY"
        order = MarketOrder(side, abs(trades))
        trade = ib.placeOrder(cfd, order)  
    elif trades < 0:
        side = "SELL"
        order = MarketOrder(side, abs(trades))
        trade = ib.placeOrder(cfd, order)
    else:
        pass
  
```

```python
def trade_reporting():
    global report
    
    fill_df = util.df([fs.execution for fs in ib.fills()])[["execId", "time", "side", "cumQty", "avgPrice"]].set_index("execId")
    profit_df = util.df([fs.commissionReport for fs in ib.fills()])[["execId", "realizedPNL"]].set_index("execId")
    report = pd.concat([fill_df, profit_df], axis = 1).set_index("time").loc[session_start:]
    report = report.groupby("time").agg({"side":"first", "cumQty":"max", "avgPrice":"mean", "realizedPNL":"sum"})
    report["cumPNL"] = report.realizedPNL.cumsum()
        
    clear_output(wait=True)
    display(df, report)
```

__Create the trading session and subscribe__

```python
# start trading session
bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='1 D',
        barSizeSetting=freq,
        whatToShow='MIDPOINT',
        useRTH=True,
        formatDate=2,
        keepUpToDate=True)

last_bar = bars[-1].date
bars.updateEvent += onBarUpdate

## stop trading session
# if the time pass the time limit that we defined, its stop the session. 
while True:
    ib.sleep(5) # check every 5 seconds
    if dt.datetime.utcnow().time() >= end_time: # if stop conditions has been met
        execute_trade(target = 0) # close open position 
        ib.cancelHistoricalData(bars) # stop stream
        ib.sleep(10)
        try:
            trade_reporting() # final reporting
        except:
            pass
        print("Session Stopped.")
        ib.disconnect()
        break
    else:
        pass
```
