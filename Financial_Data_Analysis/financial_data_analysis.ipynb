## Investment Multiple and CAGR 

__Two alternative reward metrics that are more intuitive and easier to interpret.__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
msft = pd.read_csv("msft.csv", index_col = "Date", parse_dates = ["Date"])
msft
```

__Investment Multiple__: Ending Value of 1 [Dollar] invested. <br>
Multiple = Ending Value / Initial Investment

```python
# Investment Multiple in other words:
# you invested 1$ in date `x`, what the profit that you earned from this
# 1$ at date `y` ? 5.43$  
# so, you earned X5.4  
# Each 1$ of invesment, you made 5.43$

multiple = (msft.Price[-1] / msft.Price[0])
multiple
```

__Price Increase (in %)__

```python
# The proce increse in 443% !  
(multiple - 1) * 100
```

```python
# similar/identical concept: Normalized Price with Base Value 1
msft.Price / msft.Price[0]
```

__Drawback of Investment Multiple: DoesnÂ´t take into account investment Period. Meaningful only in conjunction with Investment Period.__

__Compound Annual Growth Rate (CAGR)__: The (constant annual) rate of return that would be required for an investment to grow from its beginning balance to its ending balance, assuming the profits were reinvested at the end of each year of the investment's lifespan. (Wikipedia)

```python
# start time
start = msft.index[0]
start
```

```python
# end time
end = msft.index[-1]
end
```

```python
# number of days between start to end time
td = end - start
td
```

```python
# time in years.
td_years = td.days / 365.25
td_years
```

```python
# compute the annual growth rate 
# 0.28978 -> annual growth of 28.9%
cagr = multiple**(1 / td_years) - 1 
cagr
```

```python
# the same calculation, but from strech. 

cagr = (msft.Price[-1]/msft.Price[0])**(1/((msft.index[-1] - msft.index[0]).days / 365.25)) - 1 # long
cagr
```

```python
(1 + cagr)**(td_years) # alternative #2 to calculate multiple (cagr)
```

__-> CAGR can be used to compare Investments with different investment horizons.__

## Discrete Compounding

When the number of compounding event increse -> the return of investment may be higher. 

> For example, compounding events of months, will return higher investment than year event. 

> A discrete compounding event refers to a specific time period at which interest or investment returns are calculated and added to the principal amount. This event occurs at fixed intervals, and the compounding frequency can vary depending on the financial instrument or investment.

**Here are some key points to understand discrete compounding in finance:**

- **Compounding Frequency**: In discrete compounding, the compounding frequency specifies how often interest or returns are added to the principal. Common compounding frequencies include annually (once a year), semi-annually (twice a year), quarterly (four times a year), monthly (twelve times a year), and daily (365 times a year), among others.

__Annual Compounding__ -> Interests accrue once a year at the end of the year

Your Savings Bank offers an interest rate of __8% p.a. (stated rate)__ with __annual compounding__ on your savings (__USD 100__).<br>
Calculate the __value__ of your savings account __after one year__ and the corresponding __effective annual interest rate__. 

__-> Interests are calculated and added to your savings account once at the end of each year.__

```python
PV = 100
r = 0.08
n = 1
```

```python
# value of one year.

100 * 1.08
```

```python
# FV -> future value
# PV -> current value
# (1 + r) -> the rate in precent 
# **n -> time period. 
FV = PV * (1 + r)**n
FV
```

```python
# How many precent the investment increse per n period? 
effective_annual_rate = (FV / PV)**(1/n) - 1 
effective_annual_rate
```

## Log Returns

**Advantages of Log Returns:**

1. **Additivity**: Log returns are additive, which means that you can sum them over different time periods to calculate the cumulative return. This property is especially useful when analyzing returns over multiple periods. In contrast, simple returns are not additive, and calculating the cumulative return using simple returns can lead to inaccuracies.

2. **Normal Distribution:** Log returns tend to follow a normal distribution, which makes them suitable for many statistical and financial models that assume normally distributed returns. This simplifies modeling and analysis.

3. **Stationarity:** Log returns often exhibit better stationarity, which means that their statistical properties remain more consistent over time. This makes them more amenable to various time series analysis techniques.

4. **Percentage vs. Dollar** Changes: Log returns express returns as a percentage change in the value of an asset, which is often more informative and easier to interpret than dollar changes, especially when comparing assets with different price levels.

5. **Continuous Compounding:** Log returns are closely related to the concept of continuous compounding, which is a fundamental idea in finance. When returns are compounded continuously, the log returns closely approximate the actual compounded returns.

6. **Use in Risk and Portfolio Management:** Log returns are widely used in risk and portfolio management, including the calculation of risk measures like volatility and in the formulation of portfolio optimization models, such as the Markowitz portfolio theory.

7. **Mathematical Properties:** Logarithmic returns have desirable mathematical properties that simplify various financial calculations and modeling, such as pricing options using the Black-Scholes model.

## Normality of Financial Returns 

> Checking the distribution of a stock return, is one of the most importent thinks, becase there are lot's of statistical measure, indicators and teqnuces that assume that the data is Gasussian.

__Do MSFT Returns follow a Normal Distribution?__ <br><br>
A normally distributed random variable can be fully described by its 
- __mean__
- __standard deviation__

Higher Central Moments are zero:
- __Skew = 0__ (measures symmetrie around the mean)
- __(Excess) Kurtosis = 0__ (positve excess Kurtosis -> more observations in the "tails")

```python
# Compute the mean of the log return
mu = msft.log_ret.mean()
mu
```

```python
# Compute the std of the log return
sigma = msft.log_ret.std()
sigma
```

__Now, Lets make a few statistical measure for check if the log return is Gaussian.__

```python
import scipy.stats as stats
```

**Compute the skewness:**
- We can see that the `skew != 0`. its mean that the distribution is not symmatric around the mean, and therefor its probably not normal distribution.
    - `skew = -0.200705`, mean that there is negative assymetry. (the distribution assymetry to the negative.)



```python
# Compute the skewness. 
stats.skew(msft.log_ret.dropna()) # in a Normal Distribution: skew == 0
```

**Compute the kurtosis**
- We can see that the `kurtosis` is not 0, so its mean taht the distribution is not normal, becase there is a kurtosis. 

```python
stats.kurtosis(msft.log_ret.dropna(), fisher = True) # in a Normal Distribution: (fisher) kurtosis == 0
```

__MSFT Returns exhibit "Fat Tails" (extreme positive/negative outcomes).__

So, what we gonna to do?
- base on the `mean` and `std`, `min` & `max` value of the distribution, we are going to create the expected normal distribution, and then we are going to compare the expected distribution to the actuall distribution. 

There is two cases:
- If the expected and the actual distribution are similar, we can fill comftble that the  stocl distribution is normal.
- else, the distribution of the stock return is not normal.

```python
# 1. Create linear space with range of [min,max] log return.
x = np.linspace(msft.log_ret.min(), msft.log_ret.max(), 10000)
x
```

```python
# 2. Using the probably density function, I create the 
# expected normal distribution with the next parameters:
# x -> the linear space
# loc -> location, the expected value (the mean)
# scale -> the std.
y = stats.norm.pdf(x, loc = mu, scale = sigma) # creating y values a for normal distribution with mu, sigma
y
# now, i created the expected normal distribution. 
```

```python
# 3. Plot the expected normal distribution and the actual distribution
# for comparison.
plt.figure(figsize = (20, 8))
plt.hist(msft.log_ret, bins = 500, density = True, label = "Frequency Distribution of daily Returns (MSFT)")
plt.plot(x, y, linewidth = 3, color = "red", label = "Normal Distribution")
plt.title("Normal Distribution", fontsize = 20)
plt.xlabel("Daily Returns", fontsize = 15)
plt.ylabel("pdf", fontsize = 15)
plt.legend(fontsize = 15)
plt.show()

# We can see clearly that our distribution, is not normal.
```

__-> MSFT Returns exhibit "Fat Tails" (extreme positive/negative outcomes).__ 

**Hypotesis test - for testing whether our distribution is normally or not.**

- Testing the normality of MSFT Returns based on the sample (Oct 2014 to May 2021): <br>
- Hypothesis Test with H0 Hypothesis: MSFT Returns (full population) follow a normal Distribution. 

- `H0`: the null hypotheses; the data **is** normally distributed.
- `H1`: the althernative hypotheses; the data **is not** normaly distributed

- `z_stat`: This variable holds the Z-statistic, which is a measure of how many standard deviations the sample data is from the mean of a normal distribution. In a normality test, a Z-statistic that is significantly different from zero may indicate that the data **significantly deviates from a normal distribution.**

- `p_value`:  The p-value is a measure of the evidence against a null hypothesis. In the context of a normality test, **a small p-value** (typically less than 0.05) suggests that there is evidence to reject the null hypothesis that the data follows a normal distribution.

```python
# Make the normal test (test if its normal)
# extract the z_stat, p_value:
# z_stat 
z_stat, p_value = stats.normaltest(msft.log_ret.dropna())
```

In that case, we can see that the **z-statistics** is big, and the **p-value** is small. 

**Its mean that we reject the H0. In other words, the data DONT came from normal distribution**

```python
z_stat # high values -> reject H0
```

```python
p_value # low values (close to zero) -> reject H0
```

```python
round(p_value, 10)
```

**Conclution**
- . We checked if the distribution of stock return come from Gaussian. 
- We got that the **distribution of stock return DO NOT came from the normal distribution.**

**What meaning of that?**
- because its NOT came from noraml, we can trust on the `std` and `mean` (they are understatmant). 
- Do not expect that the stock behavior will be as normal distribution. we need to expect to exptreme events and uncerteny. 
- assume that the stock behavior as "random walk" (stock behavior randomaly)

__-> Assuming that MSFT Returns (generally) follow a Normal Distribution, there is 0% probability that we get that extreme outcomes in a sample.__

__How will the Mean-Variance Analysis change with smoothed data?__

```python
freqs = ["A", "Q", "M", "W-Fri", "D"] # [annual,quarter,monthly,weekly,daily]
periods = [1, 4, 12, 52, 252] # [1year,4 qarters in a year, and so on..]

ann_mean = []
ann_std = []
```

```python
# Compute the mean and std for each period, then store it in a list.
for i in range(5):
    # resample the freqiency 
    resamp = msft.Price.resample(freqs[i]).last()
    
    # compute the mean/std 
    period_mean_return = np.log(resamp / resamp.shift()).mean() * periods[i] 
    period_std_return = np.log(resamp / resamp.shift()).std() * np.sqrt(periods[i]) 
    
    # store the result
    ann_mean.append(period_mean_return) 
    ann_std.append(period_std_return) 
```

```python
# mean return for each period (frequency)
ann_mean
```

```python
#  store the std and mean in a df.
summary = pd.DataFrame(data = {"ann_std":ann_std, "ann_mean":ann_mean}, index = freqs)
summary
```

```python
# Plot the std and mean for each period.
summary.plot(kind = "scatter", x = "ann_std", y = "ann_mean", figsize = (15,12), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "ann_std"]+0.001, summary.loc[i, "ann_mean"]+0.001), size = 15)
plt.ylim(0, 0.3)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

## Margin Trading & Levered Returns (Part 1)

__Definition__: "Margin trading refers to the practice of using __borrowed funds__ from a broker to trade a financial asset, which forms the collateral for the loan from the broker." (Investopedia.com) 

In Simple Words: Investors __donÂ´t pay the full price__ but they get the full benefit (less borrowing costs).

ItÂ´s a two edged sword: Leverage __amplifies both gains and losses__. <br> In the event of a loss, the collateral gets reduced and the Investor either posts additional margin or the brokers closes the position.

__Example__

A Trader buys a stock (stock price: 100) __on margin (50%)__. After one day the price increases to 110.<br>
Calculate __unlevered return__ and __levered return__.

```python
P0 = 100 # price day 1
P1 = 110 # price day 2

leverage = 2 # the lavarage (2x)
margin = P0/2 
```

```python
margin
```

```python
# un-leverage return
unlev_return = (P1 - P0) / P0 # simple return
unlev_return
```

```python
# the leverage return (2x)! 
lev_return = (P1 - P0) / margin # simple return 
lev_return
```

```python
# this relationship is true for simple returns...
lev_return == unlev_return * leverage
```

```python
# un-leverage log return 
unlev_return = np.log((P1 - P0) / P0 + 1) # log return
unlev_return
```

```python
# leverage log return
lev_return = np.log((P1 - P0) / margin + 1) # log return
lev_return
```

```python
lev_return == unlev_return * leverage # this relationship does not hold for log returns...
```

__Take Home: To calculate levered returns, donÂ´t multiply leverage with log returns!__

__Coding Challenge #7__

1. Calculate levered returns for Bitcoin (leverage = 4). 

2. Visualize and compare with unlevered Investment.

3. Some Traders trade Bitcoin with extremely high leverage (> 100). Do you think this is a good idea (assuming no additional/advanced Risk Management Tools)?

```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.4f}'.format
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
close = pd.read_csv("close.csv", index_col = "Date", parse_dates = ["Date"])
close
```

```python
btc = close["BTC-USD"].dropna().to_frame().copy()
btc
```

```python
btc["Returns"] = btc.pct_change(periods = 1)
btc
```

```python
leverage = 4
```

```python
btc["Lev_Returns"] = btc.Returns.mul(leverage) # levered simple returns
btc
```

```python
btc["Lev_Returns"] = np.where(btc["Lev_Returns"] < -1, -1, btc["Lev_Returns"])
```

```python
btc[["Returns", "Lev_Returns"]].add(1).cumprod().plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

-> Trading Bitcoin with (high) leverage requires advanced risk monitoring/management. Otherwise, a complete loss is very likely (sooner or later...).
